```python
import pandas as pd
```

```python
def per_48_minutes(row, stat="BLK"):
    return row[stat] * (48 / row["MIN"])
```

```python
df = pd.read_csv("dpy_df.csv")
df
```

```python
df['BLK/48'] = df.apply(per_48_minutes, axis=1)
df["REB/48"] = df.apply(per_48_minutes, stat="REB", axis=1)
df["STL/48"] = df.apply(per_48_minutes, stat="STL", axis=1)
```

```python
df
```

```python
df_names = pd.read_csv("DPY_win.csv", sep=';')
df_names
#Columns: "Date, Name, ID
```

```python

# Now perform the join operation
result_df = df.join(df_names.set_index('ID'), on='Player_ID', how='left')
result_df
```

```python
result_df = result_df.drop(columns=["SEASON_ID", "Player_ID", "Game_ID"])
```

```python
result_df.to_csv("dpoy_winners.csv", index=False)
```
